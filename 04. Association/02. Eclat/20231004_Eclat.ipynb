```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

## Preprocesamiento de datos

```python
dataset = pd.read_csv('Market_Basket_Optimisation.csv', header = None)
transactions = []
for i in range(0, 7501):
  transactions.append([str(dataset.values[i,j]) for j in range(0, 20)])
```

## Entrenamiento del modelo Eclat en el conjunto de datos

```python
from apyori import apriori
rules = apriori(transactions = transactions, min_support = 0.003, min_confidence = 0.2, min_lift = 3, min_length = 2, max_length = 2)
```

## Visualización de los resultados

### Visualización de los primeros resultados procedentes directamente de la salida de la función apriori

```python
results = list(rules)
```

```python
results
```

### Poner los resultados bien organizados en un Pandas DataFrame

```python
def inspect(results):
    lhs         = [tuple(result[2][0][0])[0] for result in results]
    rhs         = [tuple(result[2][0][1])[0] for result in results]
    supports    = [result[1] for result in results]
    return list(zip(lhs, rhs, supports))
resultsinDataFrame = pd.DataFrame(inspect(results), columns = ['Product 1', 'Product 2', 'Support'])
```

### Mostrar los resultados ordenados por soportes descendentes

```python
resultsinDataFrame.nlargest(n = 10, columns = 'Support')
```
